```python
import numpy as np
import matplotlib.pyplot as plt

phas0 = np.loadtxt('lcls.dist',skiprows=5)

mass = 0.511         #MeV

gam0 = np.mean(phas0[:,5])
E0   = gam0*mass

z0    = -3e8*(phas0[:,4]-np.mean(phas0[:,4])) *1e6                   # um
delta = (phas0[:,5]-gam0)/gam0 * 100   # %
```

```python
plt.figure
# plt.plot(z0,delta,'.')
plt.hexbin(z0,delta,gridsize=(200,200),mincnt=1,cmap=plt.cm.jet)

plt.xlabel('z (um)')
plt.ylabel('$\delta$ %')
```
